### Looping through multiple pages

```python
def get_attribute(parent, selector, attribute):
    try:
        if attribute == 'text':
            val =  parent.find_element_by_css_selector(selector).text
        else:
            val = parent.find_element_by_css_selector(selector).get_attribute(attribute)
    except:
        val = None
    return val

import time
jobs_df = pd.DataFrame()
for i in range(1, 10):
    url = 'https://www.naukri.com/data-science-jobs-%d' % (i)
    browser.get(url)
    time.sleep(2)
    print(i)
    parents = browser.find_elements_by_css_selector('div.row[type="tuple"]')
    for job in parents:
        job_dict = {
                'id': job.get_attribute('id'),
                'url': job.get_attribute('data-url'),
                'title': get_attribute(job, 'li.desig', 'title'),
                'company': get_attribute(job, 'span.org', 'text'),
                'experience': get_attribute(job, 'span.exp', 'text'),
                'location': get_attribute(job, 'span.loc', 'text'),
                'skills': get_attribute(job, 'span.skill', 'text'),
                'salary': get_attribute(job, 'span.salary', 'text'),
                'date': get_attribute(job, 'span.date', 'text'),
            }
        jobs_df = jobs_df.append(job_dict, ignore_index=True)
```
