
# Joins

```python
# dataframe should be of the same length
df1  = pd.DataFrame({'key1': ['a', 'b','e','f'], 'val1': [1, 2,9,10]})
df2 = pd.DataFrame({'key2': ['a', 'b','c','d'], 'val2': [4, 5,6,7]})

print("df1: ")
print(df1)

print("df2")
print(df2)
```

```python
pd.merge(df2,df1, on = df1.key1)
```

```python
df1  = pd.DataFrame({'key': ['a', 'b','e','f'], 'val1': [1, 2,9,10]})
df2 = pd.DataFrame({'key': ['a', 'b','c','d'], 'val2': [4, 5,6,7]})
pd.merge(df2,df1, on = 'key')
```

# right outer join

```python
df1  = pd.DataFrame({'key1': ['a', 'b','e','f'], 'val1': [1, 2,9,10]})
df2 = pd.DataFrame({'key2': ['a', 'b','c','d'], 'val2': [4, 5,6,7]})
pd.merge(df2,df1, left_on = 'key2' , right_on = 'key1', how ="right") 
```

# left outer join

```python
df1  = pd.DataFrame({'key1': ['a', 'b','e','f'], 'val1': [1, 2,9,10]})
df2 = pd.DataFrame({'key2': ['a', 'b','c','d'], 'val2': [4, 5,6,7]})
pd.merge(df2,df1, left_on = 'key2' , right_on = 'key1', how ="left") 
```

# concate command

```python
pd.concat([df2,df1])
```

```python
pd.concat([df2,df1] , axis =0 ) # rbind
```

```python
pd.concat([df2,df1] , axis =1 ) # cbind
```
