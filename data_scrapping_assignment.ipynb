```python
import selenium
import bs4
import pandas as pd
import requests
from selenium import webdriver
```

```python
browser = webdriver.Chrome('chromedriver.exe')
url = "https://www.goodreads.com/choiceawards/best-books-2017"
browser.get(url)
```

## 1.	From this website, extract the following information in a data frame format

a)	Name of the category
b)	Book name
c)	URL to the book cover image
d)	URL to the book description (Clicking on the image, takes you to a new page. Identify this URL)


```python
book = browser.find_elements_by_css_selector("div.category.clearFix")
print(len(book))
```

```python
def get_info(job,selector, attribute):
    try:
        tag = job.find_element_by_css_selector(selector).get_attribute(attribute)
    except :
        tag = None
    return tag
df_b = pd.DataFrame(columns = ['cat','name','url','url_img'])
for b in book:
    cat = get_info(b,'h4','innerHTML')
    name = get_info(b,'img.winner','alt')
    url = get_info(b,'a','href')
    url_img = get_info(b,'img','src')
    cur_row = {'cat':cat,
               'name':name,
               'url':url,
               'url_img':url_img}
    df_b = df_b.append(cur_row,ignore_index = True)
df_b
```

```python
i=0
for url in df_b['url'].head(1):
    browser.get(url)
    i = i+1
    full_desc = browser.find_element_by_css_selector('span[itemprop="name"]').text
    print(i,full_desc)
```
