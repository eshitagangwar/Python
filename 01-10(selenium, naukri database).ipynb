```python
na = []
de = []
di = []
for item in prod:
    name = item.find_element_by_css_selector('div.product-brand')
    name = name.get_attribute('innerHTML')
    
    na.append(name)
    desc = item.find_element_by_css_selector('h4.product-product').get_attribute('innerHTML')
    de.append(desc)
    disc = item.find_element_by_css_selector('div.product-price')
    disc = disc.find_element_by_css_selector('span.product-discountPercentage').get_attribute('innerHTML')
    di.append(disc)
    
    print(name,"\t\t",desc,"\t\t",disc)
print(di[2])
```

```python
p = pd.DataFrame({"Brand":na,
                  "Description":de,
                  "Discount":di})
p
```

# NAUKRI WEBSITE

```python
browser = webdriver.Chrome('chromedriver.exe')
url = "https://www.naukri.com/data-science-jobs"
browser.get(url)
```

```python
jobs = browser.find_elements_by_css_selector('div.row')
print(len(jobs))
```

```python
jobs = browser.find_elements_by_css_selector('div.row[type ="tuple"]')
print(len(jobs))
```

```python
for job in jobs:
    title = job.find_element_by_css_selector('li.desig').get_attribute('title')
    org = job.find_element_by_css_selector('span.org').get_attribute('innerHTML')
    salary = job.find_element_by_css_selector('span.salary').get_attribute('innerHTML')
    print(title,'\t',org,'\t',salary)
    
```

## Creating an function

```python
def get_info(job,selector, attribute):
    try:
        tag = job.find_element_by_css_selector(selector).get_attribute(attribute)
    except :
        tag = None
    return tag
df_jobs = pd.DataFrame(columns = ['title','company','salary','date','location','experience','description','url'])
for job in jobs:
    title = get_info(job,'li.desig','title')
    org = get_info(job,'span.org','innerHTML')
    salary =get_info(job,'span.salary','innerHTML')
    date = get_info(job,'span.date','innerHTML')
    skill = get_info(job,'span.skill','innerHTML')
    desc = get_info(job,'span.desc','innerHTML')
    try:
        loc = job.find_element_by_css_selector('span.loc')
        loc = get_info(loc,'span','innerHTML')
    except:
        loc = None
    exp = get_info(job,'span.exp','innerHTML')
    url = get_info(job,'a.content','href')
    curr_row = {'title':title,
                'company':org,
                'salary':salary,
                'date':date,
                'location':loc,
                'experience':exp,
                'description':desc,
               'url':url}
    df_jobs = df_jobs.append(curr_row, ignore_index = True)
df_jobs
```

```python
for url in df_jobs['url'].head(3):
    print(url,'\n')
```
