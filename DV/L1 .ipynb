```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,9,10)
x
y = x**2
y
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('x-y plot')
plt.show()
```

```python
a = np.random.randint(10,50,5)
b = a/2
a,b
c = np.random.randint(10,50, 5)
d = c+np.random.randint(20,30,5)
a , b ,c , d
```

```python
fig = plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.plot(x,y,'g-*')
plt.xlabel('x')
plt.ylabel('y')
plt.title('x-y plot')
plt.subplot(2,2,2)
plt.plot(y,x, 'r--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('y-x plot')
plt.subplot(2,2,3)
plt.plot(a,b, 'b--')
plt.xlabel('a')
plt.ylabel('b')
plt.title('a-b plot')
plt.subplot(2,2,4)
plt.plot(c,d, 'm--')
plt.xlabel('c')
plt.ylabel('d')
plt.title('b-a plot')
plt.show()
#plt.tight_layout() if two graphs overlap
```

```python
fig = plt.figure() #creating fig object
axes1 = fig.add_axes([0.2,0.2,1,1])
axes2 = fig.add_axes([0.4,0.4, 0.6,0.6])
axes3 = fig.add_axes([0.9,-0.2, 0.8,0.8])
plt.xticks(np.arange(min(x), max(x)+1, 1.0)) #ticks for x axis
#plt.yticks(np.arange(min(y), max(y)+1, 5))
#plt.xticks(np.arange(min(a), max(a)+1, 5))
axes1.plot(x,y, 'g')
axes2.plot(a,b, 'r')
axes3.plot(c,d, 'b')
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2, figsize = (12,7))
axes[0].plot(x,y)
axes[1].plot(a,b, 'g')
axes[0].set_title('x-y plot')
axes[1].set_title('a-b plot')
plt.savefig('2plots.png')
plt.show()
```
