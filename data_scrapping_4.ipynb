```python
import selenium
import bs4
import pandas as pd
import requests
from selenium import webdriver
```

```python
url = "https://www.gsmarena.com/apple_iphone_6-reviews-6378.php"
```

```python
browser = webdriver.Chrome('chromedriver.exe')
browser.get(url)
```

```python
book = browser.find_elements_by_css_selector("div.user-thread")
len(book)
```

```python
# the parent tag is div.user-thread
```

```python
def get_info(job,selector, attribute):
    try:
        tag = job.find_element_by_css_selector(selector).get_attribute(attribute)
    except :
        tag = None
    return tag
df_b = pd.DataFrame(columns = ['tite','text','date','upvotes'])
for b in book:
    
    title = get_info(b,'li.uname2','innerHTML')
    #print(title)
    
    text = get_info(b,'p.uopin','innerHTML')
    #print(text)
    date = get_info(b,'li.upost','innerHTML')
    #print(date)
    
    upvotes = get_info(b,'span.uavatar-stat.upvote-count','innerHTML')
    #print(upvotes)
    cur_row = {'tite':title,
               'text':text,
               'date':date,
               'upvotes':upvotes}
    df_b = df_b.append(cur_row,ignore_index = True)
df_b
```
